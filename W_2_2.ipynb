### Derin Öğrenme

```python
from sklearn. datasets import load_iris
import numpy as np
import pandas as pd
```

```python
data = load_iris ()
X = data.data
y = data.target

input_dim = X.shape[1] # Sutun sayisi (Ozellikler)

output_dim = len(np.unique(y))

print(f"Giris Noronu Sayisi: {input_dim}") # Cikti: 4
print(f"Cikis Noronu Sayisi: {output_dim}") # Cikti: 3
```
