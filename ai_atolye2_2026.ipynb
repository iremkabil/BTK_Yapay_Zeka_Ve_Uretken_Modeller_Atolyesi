## 1. Makine Öğrenmesine Giriş

**Makine öğrenmesi**, sistemlerin açıkça programlanmaksızın veri üzerinden öğrenmesini sağlayan bir yapay zeka alt alanıdır.

| | Klasik Programlama | Makine Öğrenmesi |
|---|---|---|
| **Girdi** | Kurallar + Veri | Veri + Cevaplar (Etiketli Veri) |
| **Çıktı** | Sonuç | Kurallar (Model) |

### Öğrenme Türleri
- **Gözetimli Öğrenme (Supervised):** Etiketli veri → Sınıflandırma, Regresyon
- **Gözetimsiz Öğrenme (Unsupervised):** Etiketsiz veri → Kümeleme, Boyut İndirgeme
- **Pekiştirmeli Öğrenme (Reinforcement):** Ajan + Çevre → Ödül/Ceza döngüsü
- **Yarı Gözetimli (Semi-Supervised):** Az etiketli + çok etiketsiz veri

### ML Lifecycle (Modelleme Döngüsü)
```
Veri Toplama → Veri Hazırlama → Özellik Seçimi → Model Seçimi
     ↑                                                   ↓
  İyileştirme ← Değerlendirme ←──────── Model Eğitimi
```

## 4. Eksik Veri Yönetimi (Missing Data Imputation)

````python
import pandas as pd
from sklearn.impute import SimpleImputer

# Örnek veri seti oluşturma
df = pd.DataFrame({
    'yas':  [25, np.nan, 35, 40, np.nan, 50, 28, 33],
    'maas': [3000, 4500, np.nan, 5500, 3800, np.nan, 4200, 5100],
    'sehir': ['Istanbul', 'Ankara', np.nan, 'Izmir', 'Istanbul', 'Ankara', np.nan, 'Bursa']
})

print('Orijinal Veri:')
print(df)
print('\nEksik veri sayısı:')
print(df.isnull().sum())
````

````python
# 1. Yöntem: Pandas ile hızlı doldurma (medyan ile)
df_filled = df.copy()
df_filled['yas'] = df_filled['yas'].fillna(df_filled['yas'].median())

# 2. Yöntem: Scikit-Learn SimpleImputer (sayısal)
imputer = SimpleImputer(strategy='median')
df_filled[['maas']] = imputer.fit_transform(df_filled[['maas']])

# 3. Yöntem: Kategorik eksik veri — en sık değer ile doldurma
df_filled['sehir'] = df_filled['sehir'].fillna(df_filled['sehir'].mode()[0])

print('Doldurulmuş Veri:')
print(df_filled)
print('\nKalan eksik veri:')
print(df_filled.isnull().sum())
````

### Eksik Veri Kararı: Silmek mi, Doldurmak mı?

| Durum | Öneri |
|---|---|
| Veri seti küçük & çok eksik var | **Doldurma** tercih edilir |
| Eksik değişken model için önemsiz | **Sütunu sil** |
| Kritik değişkende eksik | **Model tabanlı doldurma** (KNN/Regresyon) |
| Eksiklik oranı %1-2 | O **satırları sil** |

## 5. Özellik Ölçekleme (Feature Scaling)

**⚠️ Veri Sızıntısı (Data Leakage) Önlemi:**  
`fit()` işlemi yalnızca **eğitim verisi** üzerinde yapılmalıdır. Test verisine sadece `transform()` uygulanır.

````python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Örnek veri
X_ornek = pd.DataFrame({
    'yas':  [20, 35, 50, 25, 45],
    'maas': [2000, 5000, 8000, 3000, 7000]
})

# Z-score standardizasyonu (Ortalama=0, Varyans=1)
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X_ornek)

# Min-Max normalizasyonu (0-1 aralığı)
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X_ornek)

print('Orijinal Veri:')
print(X_ornek)
print('\nStandardScaler (Z-score):')
print(pd.DataFrame(X_std, columns=X_ornek.columns).round(3))
print('\nMinMaxScaler (0-1):')
print(pd.DataFrame(X_minmax, columns=X_ornek.columns).round(3))
````

## 6. Etiket Kodlama (Encoding)

Makine öğrenmesi modelleri metinsel verilerle çalışamaz. Kategorik verileri sayısal forma dönüştürmek gerekir.

- **Label Encoding:** Sıralı (ordinal) veriler için (İlkokul < Lise < Üniversite)
- **One-Hot Encoding:** Sırasız (nominal) veriler için (Kırmızı, Mavi, Yeşil)

````python
from sklearn.preprocessing import LabelEncoder

# ---- Label Encoding (Sıralı veri için) ----
df_egitim = pd.DataFrame({'egitim': ['Lise', 'Ilkokul', 'Universite', 'Lise']})

encoder = LabelEncoder()
df_egitim['egitim_kodlu'] = encoder.fit_transform(df_egitim['egitim'])

print('Label Encoding (Eğitim Durumu):')
print(df_egitim)
print('\nSınıflar:', encoder.classes_)
````

````python
# ---- One-Hot Encoding (Sırasız nominal veri için) ----
df_renk = pd.DataFrame({'renk': ['Kirmizi', 'Mavi', 'Yesil', 'Mavi']})

# drop_first=True: Dummy değişken tuzağını önler
# Eğer Kirmizi ve Mavi 0 ise → otomatik olarak Yesil anlaşılır
onehot_df = pd.get_dummies(df_renk, columns=['renk'], drop_first=True, dtype=int)

print('One-Hot Encoding (Renk):')
print(onehot_df)
print('\nNot: İlk sütun (Kirmizi) drop_first=True ile çıkarıldı.')
print('Her iki sütun da 0 ise renk Kirmizi anlaşılır.')
````

## 7. Veri Bölme (Train/Test Split)

**Altın Kural:** Test verisi eğitim aşamasında **ASLA** kullanılmaz!

- Yaygın oranlar: **80/20** veya **70/30**
- `stratify=y`: Sınıf dağılımını korur (dengesiz veri için kritik)

````python
from sklearn.model_selection import train_test_split

# Hedef (y) dağılımı: %90 Sınıf 0, %10 Sınıf 1
y_imbalanced = [0] * 90 + [1] * 10
X_imbalanced = list(range(100))

# Yanlış Yaklaşım: Rastgele bölme (Test setine hiç 1 gelmeyebilir!)
_, _, y_train_bad, y_test_bad = train_test_split(
    X_imbalanced, y_imbalanced, test_size=0.2, random_state=0
)

# Stratify ile orantılı bölme
_, _, y_train_good, y_test_good = train_test_split(
    X_imbalanced, y_imbalanced, test_size=0.2, stratify=y_imbalanced, random_state=42
)

print('=== Rastgele Bölme (Yanlış) ===')
print(f'Test setinde Sınıf 1 sayısı: {sum(y_test_bad)}')
print(f'\n=== Stratify ile Bölme (Doğru) ===')
print(f'Test setinde Sınıf 1 sayısı: {sum(y_test_good)} (Beklenen ~2)')
````

## 10. Lojistik Regresyon (Logistic Regression)

İkili sınıflandırma (Binary Classification) algoritmasıdır. **Sigmoid Fonksiyonu** ile doğrusal denklemin çıktısını 0-1 arasına sıkıştırır:

$$\hat{y} = \frac{1}{1 + e^{-(wX+b)}}$$

**Karar Sınırı:** Olasılık > 0.5 → Sınıf 1, < 0.5 → Sınıf 0

````python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Örnek veri üret
X_cls, y_cls = make_classification(n_samples=200, n_features=2, n_redundant=0,
                                    random_state=42)
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42
)

# Lojistik Regresyon
model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train_cls, y_train_cls)
y_pred_lr = model_lr.predict(X_test_cls)

print(f'Lojistik Regresyon Doğruluğu: {accuracy_score(y_test_cls, y_pred_lr):.4f}')

# Sigmoid fonksiyonunu görselleştir
x_sig = np.linspace(-6, 6, 200)
y_sig = 1 / (1 + np.exp(-x_sig))

plt.figure(figsize=(8, 4))
plt.plot(x_sig, y_sig, 'b-', linewidth=2, label='Sigmoid')
plt.axhline(0.5, color='r', linestyle='--', label='Karar Sınırı (0.5)')
plt.axvline(0, color='gray', linestyle=':')
plt.xlabel('z = wX + b')
plt.ylabel('Olasılık')
plt.title('Sigmoid Fonksiyonu')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 11. K-En Yakın Komşu (KNN)

**Lazy Learning (Tembel Öğrenme):** Eğitim aşamasında model kurulmaz; hesaplama tahmin anında yapılır.

- Sınıflandırma → **Çoğunluk Oylaması**
- Regresyon → **Komşuların Ortalaması**
- ⚠️ K küçükse → **Overfitting**, K büyükse → **Underfitting**

````python
from sklearn.neighbors import KNeighborsClassifier

# Farklı K değerleriyle KNN
k_values = [1, 3, 5, 7, 11]
knn_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_cls, y_train_cls)
    score = accuracy_score(y_test_cls, knn.predict(X_test_cls))
    knn_scores.append(score)
    print(f'K={k:2d} → Doğruluk: {score:.4f}')

plt.figure(figsize=(7, 4))
plt.plot(k_values, knn_scores, 'o-', color='teal', linewidth=2)
plt.xlabel('K değeri')
plt.ylabel('Doğruluk Skoru')
plt.title('KNN: K Parametresinin Etkisi')
plt.xticks(k_values)
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 12. Karar Ağacı (Decision Tree)

Veri setini **if/else kurallarıyla** dallara böler. En yüksek **Bilgi Kazancı (Information Gain)** olan özelliğe göre bölünür.

- ✅ Yorumlanabilir (White Box), ölçekleme gerektirmez
- ⚠️ Overfitting'e çok müsait → `max_depth` ile sınırlandır

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Karar Ağacı - derinlik sınırlandırılmış
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train_cls, y_train_cls)
y_pred_dt = dt.predict(X_test_cls)

print(f'Karar Ağacı (max_depth=3) Doğruluğu: {accuracy_score(y_test_cls, y_pred_dt):.4f}')

# Ağaç görselleştirmesi
plt.figure(figsize=(14, 6))
plot_tree(dt, filled=True, feature_names=['Özellik 1', 'Özellik 2'],
          class_names=['Sınıf 0', 'Sınıf 1'], rounded=True, fontsize=9)
plt.title('Karar Ağacı Görselleştirmesi (max_depth=3)')
plt.tight_layout()
plt.show()
````

## 13. Random Forest (Rastgele Orman)

**Ensemble Learning (Topluluk Öğrenmesi):** Veri setinin farklı alt kümelerini (Bagging/Bootstrap) kullanarak **yüzlerce Karar Ağacı** eğitir.

> *"Hastalığı tek bir doktora sormak yerine yüzlerce doktordan oluşan bir kurula danışmak."*

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_cls, y_train_cls)
y_pred_rf = rf.predict(X_test_cls)

print(f'Random Forest Doğruluğu: {accuracy_score(y_test_cls, y_pred_rf):.4f}')

# Özellik önem sıralaması
importances = pd.Series(rf.feature_importances_,
                         index=['Özellik 1', 'Özellik 2'])
print('\nÖzellik Önem Sıralaması:')
print(importances.sort_values(ascending=False))
````

## 14. Destek Vektör Makineleri (SVM)

Sınıfları birbirinden ayıran **maksimum marjlı hiperdüzlemi** (Hyperplane) çizer.

**Kernel Trick:** Lineer ayrılmayan verileri daha yüksek boyuta taşıyarak ayırır.

- ⚠️ **Ölçekleme zorunludur!** StandardScaler kullanılmalıdır.

````python
from sklearn.svm import SVC

# SVM için ölçekleme zorunlu!
scaler_svm = StandardScaler()
X_train_svm = scaler_svm.fit_transform(X_train_cls)
X_test_svm = scaler_svm.transform(X_test_cls)

svm = SVC(kernel='rbf', C=1.0, random_state=42)
svm.fit(X_train_svm, y_train_cls)
y_pred_svm = svm.predict(X_test_svm)

print(f'SVM (RBF Kernel) Doğruluğu: {accuracy_score(y_test_cls, y_pred_svm):.4f}')
print(f'Destek Vektör Sayısı: {svm.n_support_}')
````

## 16. Sınıflandırma Metrikleri

### Karmaşıklık Matrisi (Confusion Matrix)

|  | Tahmin: Pozitif | Tahmin: Negatif |
|---|---|---|
| **Gerçek: Pozitif** | TP (Doğru Pozitif) | FN (Yanlış Negatif - Tip II Hata) |
| **Gerçek: Negatif** | FP (Yanlış Pozitif - Tip I Hata) | TN (Doğru Negatif) |

- **FN (False Negative):** Tıbbi teşhiste kritik → *"Hastaya sağlıklısın demek"*
- **FP (False Positive):** Spam filtrede kullanıcıyı rahatsız eder → *"Önemli maili çöpe atmak"*

### Metrikler:
- **Accuracy:** Doğru tahmin / Toplam tahmin
- **Precision:** TP / (TP + FP) → *"Spam dediklerim gerçekten spam mı?"*
- **Recall:** TP / (TP + FN) → *"Hastalıkları yakalayabildim mi?"*
- **F1 Skoru:** 2 × (Precision × Recall) / (Precision + Recall)

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

# Lojistik Regresyon sonuçları üzerinden metrikleri hesapla
print('=== Sınıflandırma Metrikleri (Lojistik Regresyon) ===')
print(f'Accuracy  : {accuracy_score(y_test_cls, y_pred_lr):.4f}')
print(f'Precision : {precision_score(y_test_cls, y_pred_lr):.4f}')
print(f'Recall    : {recall_score(y_test_cls, y_pred_lr):.4f}')
print(f'F1 Score  : {f1_score(y_test_cls, y_pred_lr):.4f}')

print('\n=== Classification Report ===')
print(classification_report(y_test_cls, y_pred_lr))
````

````python
# Confusion Matrix Görselleştirmesi
cm = confusion_matrix(y_test_cls, y_pred_lr)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Tahmin: 0', 'Tahmin: 1'],
            yticklabels=['Gerçek: 0', 'Gerçek: 1'])
plt.title('Confusion Matrix (Karmaşıklık Matrisi)')
plt.tight_layout()
plt.show()
````

## 17. Aşırı Öğrenme (Overfitting) ve Yetersiz Öğrenme (Underfitting)

| | Eğitim Başarısı | Test Başarısı | Sorun |
|---|---|---|---|
| **Underfitting** | Düşük | Düşük | Model çok basit |
| **Optimum Fit** | Yüksek | Yüksek | İdeal |
| **Overfitting** | Çok Yüksek | Düşük | Model ezberlemiş |

````python
# Karar Ağacı Derinliği vs. Overfitting Demonstrasyonu
from sklearn.datasets import make_classification

X_ov, y_ov = make_classification(n_samples=300, n_features=5, random_state=0)
X_tr, X_te, y_tr, y_te = train_test_split(X_ov, y_ov, test_size=0.3, random_state=0)

depths = range(1, 20)
train_scores, test_scores = [], []

for d in depths:
    dt_d = DecisionTreeClassifier(max_depth=d, random_state=0)
    dt_d.fit(X_tr, y_tr)
    train_scores.append(accuracy_score(y_tr, dt_d.predict(X_tr)))
    test_scores.append(accuracy_score(y_te, dt_d.predict(X_te)))

plt.figure(figsize=(9, 5))
plt.plot(depths, train_scores, 'b-o', label='Eğitim Skoru', markersize=5)
plt.plot(depths, test_scores, 'r-o', label='Test Skoru', markersize=5)
plt.axvline(x=test_scores.index(max(test_scores)) + 1,
            color='green', linestyle='--', label='Optimum Derinlik')
plt.xlabel('Ağaç Derinliği (max_depth)')
plt.ylabel('Doğruluk Skoru')
plt.title('Overfitting vs. Underfitting — Karar Ağacı Derinliği')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

opt_depth = test_scores.index(max(test_scores)) + 1
print(f'En iyi test skoru: {max(test_scores):.4f} (max_depth={opt_depth})')
````

## 19. Uygulama 1: Diabetes — Regresyon

**Amaç:** 442 hastanın 10 ölçümüne bakarak 1 yıl sonraki hastalık ilerleyişini tahmin etmek.

**Regresyon Metrikleri:**
- **MSE:** Hataların karesinin ortalaması (aykırı değere duyarlı)
- **MAE:** Hataların mutlak değerinin ortalaması (yorumu kolay)
- **R²:** Modelin açıkladığı değişkenlik oranı (1'e yakın → iyi)

````python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

# Veri yükleme
diabetes = load_diabetes()
X_diab = diabetes.data
y_diab = diabetes.target

print(f'Veri boyutu: {X_diab.shape}  →  ({X_diab.shape[0]} hasta, {X_diab.shape[1]} özellik)')
print(f'Özellikler: {diabetes.feature_names}')
print(f'Hedef (hastalık ilerleyişi): min={y_diab.min():.0f}, max={y_diab.max():.0f}, ort={y_diab.mean():.1f}')
````

````python
# BMI ile hedef arasındaki ilişki (Özellik-Hedef görselleştirmesi)
plt.figure(figsize=(7, 5))
plt.scatter(X_diab[:, 2], y_diab, alpha=0.6, color='steelblue')
plt.xlabel('BMI (Vücut Kitle İndeksi)')
plt.ylabel('Hedef (Hastalık İlerleyişi)')
plt.title('BMI vs. Hastalık İlerleyişi')
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# Eğitim ve test bölünmesi
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_diab, y_diab, test_size=0.2, random_state=42
)

# Model oluştur ve eğit
model_diab = LinearRegression()
model_diab.fit(X_train_d, y_train_d)

# Tahmin yap
y_pred_d = model_diab.predict(X_test_d)

# Metrikleri hesapla
mse = mean_squared_error(y_test_d, y_pred_d)
mae = mean_absolute_error(y_test_d, y_pred_d)
r2  = r2_score(y_test_d, y_pred_d)

print('=== Regresyon Değerlendirme Metrikleri ===')
print(f'Mean Squared Error  (MSE): {mse:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'R² Skoru                 : {r2:.4f}')
print(f'\nYorum: Model, hedef değişkendeki varyasyonun %{r2*100:.1f}\'ini açıklıyor.')
````

````python
# Gerçek vs. Tahmin görselleştirmesi
plt.figure(figsize=(10, 5))
plt.plot(y_test_d[:30], 'b-o', label='Gerçek', markersize=5)
plt.plot(y_pred_d[:30], 'r-x', label='Tahmin', markersize=7, linewidth=1.5)
plt.xlabel('Örnek No')
plt.ylabel('Skor')
plt.title('Gerçek vs. Tahmin (İlk 30 Örnek)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 20. Uygulama 2: Iris — Kümeleme Karşılaştırması

````python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA

# Iris veri seti
iris = load_iris()
X_iris = iris.data
y_iris_true = iris.target

print(f'Iris veri boyutu: {X_iris.shape}')
print(f'Sınıflar: {iris.target_names}')
print(f'Özellikler: {iris.feature_names}')
````

````python
# PCA ile 2 boyuta indirgeme (görselleştirme için)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_iris)

print(f'Açıklanan Varyans Oranı: %{sum(pca.explained_variance_ratio_)*100:.1f}')
````

````python
# Kümeleme algoritmalarını uygula
kmeans  = KMeans(n_clusters=3, random_state=0, n_init='auto')
dbscan  = DBSCAN(eps=0.6, min_samples=4)
agg     = AgglomerativeClustering(n_clusters=3)

labels_kmeans = kmeans.fit_predict(X_iris)
labels_dbscan = dbscan.fit_predict(X_iris)
labels_agg    = agg.fit_predict(X_iris)

print(f'KMeans   - Benzersiz kümeler: {np.unique(labels_kmeans)}')
print(f'DBSCAN   - Benzersiz kümeler: {np.unique(labels_dbscan)} (-1 = gürültü)')
print(f'Agg.Clust- Benzersiz kümeler: {np.unique(labels_agg)}')
````

````python
# 4 grafik: Gerçek + 3 Algoritma
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

configs = [
    (y_iris_true,   'Gerçek Etiketler',          'Set1'),
    (labels_kmeans, 'KMeans (k=3)',               'Set2'),
    (labels_dbscan, 'DBSCAN (eps=0.6, min_s=4)', 'tab10'),
    (labels_agg,    'Agglomerative (k=3)',        'Set3'),
]

for ax, (labels, title, cmap) in zip(axes, configs):
    scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], c=labels,
                         cmap=cmap, s=30, alpha=0.8)
    ax.set_title(title, fontsize=10)
    ax.set_xlabel('PCA 1')
    ax.set_ylabel('PCA 2')
    ax.grid(True, alpha=0.3)

plt.suptitle('Iris Veri Seti — Kümeleme Algoritmaları Karşılaştırması',
             fontsize=12, y=1.02)
plt.tight_layout()
plt.show()
````

## 21. Uygulama 3: Breast Cancer — Sınıflandırma Karşılaştırması

**Amaç:** 30 hücre özelliğine bakarak tümörün **iyi huylu (benign)** ya da **kötü huylu (malignant)** olduğunu tahmin etmek.

````python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Veri yükleme
data_bc = load_breast_cancer()
X_bc = data_bc.data
y_bc = data_bc.target

print(f'Veri boyutu: {X_bc.shape}')
print(f'Sınıf dağılımı: {dict(zip(data_bc.target_names, np.bincount(y_bc)))}')
````

````python
# Veri bölme (stratify ile sınıf dağılımı korunur)
X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(
    X_bc, y_bc, test_size=0.2, stratify=y_bc, random_state=42
)

# Ölçekleme — fit sadece eğitim verisine!
scaler_bc = StandardScaler()
X_train_bc_scaled = scaler_bc.fit_transform(X_train_bc)
X_test_bc_scaled  = scaler_bc.transform(X_test_bc)

print(f'Eğitim seti: {X_train_bc_scaled.shape}')
print(f'Test seti  : {X_test_bc_scaled.shape}')
````

````python
# Modeller
models_bc = {
    'Lojistik Regresyon': LogisticRegression(max_iter=1000),
    'Karar Ağacı':        DecisionTreeClassifier(random_state=42),
    'Random Forest':      RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM':                SVC(kernel='rbf', random_state=42),
    'K-En Yakın Komşu':   KNeighborsClassifier(n_neighbors=5)
}

# Tüm modelleri eğit ve değerlendir
results_bc = []

for name, model in models_bc.items():
    model.fit(X_train_bc_scaled, y_train_bc)
    y_pred_bc = model.predict(X_test_bc_scaled)
    acc = accuracy_score(y_test_bc, y_pred_bc)
    f1  = f1_score(y_test_bc, y_pred_bc)
    results_bc.append({'Model': name, 'Accuracy': acc, 'F1 Skoru': f1})
    print(f'{name:<25} → Accuracy: {acc:.4f} | F1: {f1:.4f}')

results_df_bc = pd.DataFrame(results_bc).sort_values('Accuracy', ascending=False)
print('\n=== Sıralama ===')
print(results_df_bc.to_string(index=False))
````

````python
# Performans grafiği
fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(len(results_df_bc))
width = 0.35

bars1 = ax.barh(x + width/2, results_df_bc['Accuracy'], width, label='Accuracy', color='steelblue')
bars2 = ax.barh(x - width/2, results_df_bc['F1 Skoru'], width, label='F1 Skoru', color='coral')

ax.set_yticks(x)
ax.set_yticklabels(results_df_bc['Model'])
ax.set_xlabel('Skor')
ax.set_title('Breast Cancer — Sınıflandırma Modelleri Performans Karşılaştırması')
ax.set_xlim(0.88, 1.01)
ax.legend()
ax.grid(True, axis='x', alpha=0.4)

# Değerleri bar üzerine yaz
for bar in bars1:
    ax.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)
for bar in bars2:
    ax.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)

plt.tight_layout()
plt.show()
````

````python
# En iyi modelin Confusion Matrix'i (SVM)
best_model_name = results_df_bc.iloc[0]['Model']
best_model      = models_bc[best_model_name]
y_pred_best     = best_model.predict(X_test_bc_scaled)

cm_best = confusion_matrix(y_test_bc, y_pred_best)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Greens',
            xticklabels=data_bc.target_names,
            yticklabels=data_bc.target_names)
plt.title(f'Confusion Matrix — {best_model_name} (En İyi Model)')
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek')
plt.tight_layout()
plt.show()

print(f'\n=== {best_model_name} — Classification Report ===')
print(classification_report(y_test_bc, y_pred_best, target_names=data_bc.target_names))
````
