```python
import os
from pathlib import Path
from collections import Counter

ROOT = Path(r"C:\Users\USER\Desktop\BTK_Yapay_Zeka_Ve_Uretken_Modeller_Atolyesi\tomato")

print("Ana klasör:", ROOT)
print("Var mı?:", ROOT.exists())
print("İçerik:", [p.name for p in ROOT.iterdir()])
```

```python
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

def count_images_in_split(split_path):
    class_counts = {}
    
    if not split_path.exists():
        return class_counts
    
    for class_dir in split_path.iterdir():
        if class_dir.is_dir():
            n_images = sum(1 for f in class_dir.rglob("*") if f.suffix.lower() in IMG_EXTS)
            class_counts[class_dir.name] = n_images
    return class_counts

train_dir = ROOT / "train"
val_dir   = ROOT / "val"

train_counts = count_images_in_split(train_dir)
val_counts   = count_images_in_split(val_dir)

print("TRAIN sınıf sayısı:", len(train_counts))
print("VAL sınıf sayısı  :", len(val_counts))
print("\nTRAIN dağılımı:")
for k, v in sorted(train_counts.items()):
    print(f"{k}: {v}")

print("\nVAL dağılımı:")
for k, v in sorted(val_counts.items()):
    print(f"{k}: {v}")
```
